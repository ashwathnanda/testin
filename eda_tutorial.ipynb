Replacing values in a column using .replace

```python
d = {'no' : False , 'yes' : True}
df.replace({'voice mail plan' : d})
```

```python
columns_to_show = ['total day minutes', 'total eve minutes','total night minutes']
df.groupby(by = 'international plan')[columns_to_show].describe(percentiles = [])
```

```python
columns_to_show = ['total day minutes', 'total eve minutes','total night minutes']
df.groupby(by = 'international plan')[columns_to_show].agg([np.max,np.min,np.mean,np.std])
```

we can create summary table to see the realtions between two variables . We can also normalize it with attribute normalize = True

```python
pd.crosstab(df['churn'],
            df['voice mail plan'], 
            margins = True)
```

```python
df.pivot_table(['total day calls', 'total eve calls', 'total night calls'],
               ['area code'], aggfunc='mean')
```

Inserting column within a dataframe without using intermediate series

```python
df['total calls'] = df['total day calls'] + df['total eve calls'] + df['total intl calls'] + df['total night calls']
df.head()
```

Dropping column using drop command. Inplace makes sure that the change is within the specified dataframe. if inplace command is not used then it'll return a new dataframe after performing the operation

```python
df.drop(['total calls'], 
        axis = 1,
        inplace = True)
df.head()
```
